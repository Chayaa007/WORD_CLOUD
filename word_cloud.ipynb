```python
chunk_2 = text.copy()
```

```python
chunk_2 = [line.strip() for line in chunk_2]
```

```python
print(chunk_2)
```

```python
chunk_2_pasted = ' '.join(chunk_2)
```

```python
chunk_2_pasted
```

```python
clean_data1 = chunk_2_pasted.lower()
```

```python
clean_data2 = re.sub(r'[^\w\s]','',clean_data1) #cleaning punctuations, pattern = '\w'
```

```python
clean_data2
```

```python
clean_data3 = re.sub(r'\d+',' ', clean_data2) #cleaning digits, pattern='\d'
```

```python
# StopWords
```

```python
stopwords = set(stopwords.words('english'))
```

```python
# We will remove stopwords and seperate the words
```

```python
tokens = word_tokenize(clean_data3)
```

```python
tokens
```

```python
clean_data4 = [i for i in tokens if not i in stopwords]
```

```python
clean_data4
```

```python
clean_data5 = ' '.join(clean_data4)
```

```python
clean_data5
```

```python
len(clean_data5)
```

```python
clean_data6 = ' '.join([str(x) for x in clean_data4])
```

```python
clean_data6
```

```python
len(clean_data6)
```

```python
clean_data6 = clean_data6.strip()
```

```python
# Frequency of Words
```

```python
words_dict = {}
for word in clean_data6.split():
    words_dict[word] = words_dict.get(word, 0)+1
for key in sorted(words_dict):
    print("{}:{}".format(key,words_dict[key]))
```
